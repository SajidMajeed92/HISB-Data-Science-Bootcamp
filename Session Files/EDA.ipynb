```python
def category(price):
    if price > car.Price.median():
        return "Expensive"
    else: 
        return "Cheap"
```

```python
# apply Required a function
car['Category']= car.Price.apply(category)
```

```python
car
```

```python
# Lamda Function
car['Category_by_lambda']= car.Price.apply (lambda p: "Expensive" if p>car.Price.median() else "Cheap")
```
