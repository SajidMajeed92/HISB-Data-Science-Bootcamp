```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Clean_Car_data' ,index_col=0)
```

```python
data
```

```python
data = data[data['Price']<5000000]
data.info()
```

```python
data.describe()
```

```python
data.year = data.year.str.replace('-', '').astype('int64')
```

```python
data.info()
```

```python
X =  data[['name','company','year', 'kms_driven', 'fuel_type' ]]
X
```

```python
y  = data.Price
y
```

```python
y[886]
```

# Apply Training & Testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer 
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
oho = OneHotEncoder()
oho.fit(X[['name','company','fuel_type']])
```

# creating a column transformer  to transform categorical columns

```python
column_tansformer  = make_column_transformer((OneHotEncoder(categories = oho.categories_),['name','company','fuel_type']),
                                             remainder = 'passthrough')
```

# Make a Model

```python
lr = LinearRegression()
```

# Make a Pipeline

```python
pipe = make_pipeline(column_tansformer,lr)
```

# Fitting the Model

```python
pipe.fit(X_train,y_train)
```

```python
y_pred= pipe.predict(X_test)
```

```python
pd.DataFrame({'y_train':y_test, 'y_pred':y_pred})
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test,y_pred)
```

```python
y_pred= pipe.predict(X_train)
```

```python
pd.DataFrame({'y_train':y_train, 'y_pred':y_pred})
```

```python
r2_score(y_train,y_pred)
```

```python
mean_squared_error(y_train,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_tansformer,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```
